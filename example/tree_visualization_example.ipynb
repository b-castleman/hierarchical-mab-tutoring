```python
from graphviz import Digraph

# Plotting helper functions
def create_graph (tree, data):
    graph = Digraph(format='png', strict = True)
    for problem_name, problem_data in data.items():
        graph.node(problem_name, problem_data)
        tree.add_edges_to_progression(graph)
    return graph

def render_save_graph(graph, save_name):
    graph.render(save_name, view=False)
```

```python
# Initialize algorithm and obtain required data
from hierarchical_progression_algorithm.Joint_Progression_Algorithm import Joint_Progression_Algorithm

jpa = Joint_Progression_Algorithm("search")

os.chdir('./example') # moving back for file saving organization
```

```python
# Save the concept tree
progression_tree = jpa.progression_tree_concept
all_concepts = jpa.progression_tree_concept.all_concepts
concept_tree_save_name = "concept_tree_search"

concept_names = {concept:concept for concept in all_concepts}
progression_graph = create_graph(progression_tree, concept_names)
progression_tree.save_tree(concept_tree_save_name + '.txt')
render_save_graph(progression_graph, concept_tree_save_name + '.gv')
```

```python
# Visualize the concept tree
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img = mpimg.imread(concept_tree_save_name + '.gv.png')
imgplot = plt.imshow(img)
plt.title("Concept Tree Structure")
plt.show()
```

```python
# Save an arbitrary problem tree
concept_name = "simple_search"
progression_tree = jpa.progression_tree_problem_dict[concept_name]
all_concepts = jpa.progression_tree_problem_dict[concept_name].all_concepts
problem_tree_save_name = 'problem_tree_simple_search'

concept_names = {concept:concept for concept in all_concepts}
progression_graph = create_graph(progression_tree, concept_names)
progression_tree.save_tree(problem_tree_save_name + '.txt')
render_save_graph(progression_graph, problem_tree_save_name + '.gv')
```

```python
# Visualize the problem tree (note the question id hashes are depicted)
img = mpimg.imread(problem_tree_save_name + '.gv.png')
imgplot = plt.imshow(img)
plt.title("Problem Tree Structure")
plt.show()
```
